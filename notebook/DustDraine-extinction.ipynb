```python
sys.path.append('../python')
from dust_draine import DustDraine

d = DustDraine()
fig, axes = plt.subplots(3, 1, figsize=(12, 12))
for k, df in d.dfa.items():
    plt.sca(axes[0])
    plt.loglog(df.lwav, df.kappa_abs, label=k)
    plt.sca(axes[1])
    plt.loglog(df.lwav, df.Cext, label=k)
    plt.sca(axes[2])
    plt.semilogx(df.lwav, df.albedo, label=k)

plt.sca(axes[0])
plt.legend()
plt.xlim(1e-3, 1e3)
plt.ylim(1e-1, 2e3)
plt.xlabel('wavelength ' + r'$[\mu{\rm m}]$')
plt.ylabel('absorption cross section [cm^2/g]')

plt.sca(axes[1])
plt.legend()
plt.xlim(1e-3, 1e3)
plt.ylim(1e-25, 5e-21)
plt.xlabel('wavelength ' + r'$[\mu{\rm m}]$')
plt.ylabel(r'$C_{\rm ext}$ [cm^2/H]')

plt.sca(axes[2])
plt.xlim(1e-3, 1e3)
plt.ylim(0, 1)
plt.xlabel('wavelength ' + r'$[\mu{\rm m}]$')
plt.ylabel('albedo')
```
